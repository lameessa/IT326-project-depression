```python
import pandas as pd
import matplotlib.pyplot as plt

# حساب عدد الميزات قبل الحذف
features_before = df.shape[1]

# تقليل العتبة للارتباط العالي إلى 0.8 مثلاً
threshold = 0.8

# حساب مصفوفة الارتباط
correlation_matrix = df.corr()

# رسم مصفوفة الارتباط الحرارية
plt.figure(figsize=(10, 8))
plt.matshow(correlation_matrix, fignum=1)
plt.xticks(range(len(correlation_matrix.columns)), correlation_matrix.columns, rotation=90)
plt.yticks(range(len(correlation_matrix.columns)), correlation_matrix.columns)
plt.colorbar()
plt.show()

# تحديد الميزات المرتبطة بشكل كبير
correlated_features = set()
for i in range(len(correlation_matrix.columns)):
    for j in range(i):
        if abs(correlation_matrix.iloc[i, j]) > threshold:
            colname = correlation_matrix.columns[i]
            correlated_features.add(colname)

# حذف الميزات المرتبطة بشكل كبير
df_after = df.drop(labels=correlated_features, axis=1)

# حساب عدد الميزات بعد الحذف
features_after = df_after.shape[1]

# طباعة النتائج
print(f"عدد الميزات قبل الحذف: {features_before}")
print(f"عدد الميزات بعد الحذف: {features_after}")
print(f"الميزات التي تم حذفها: {correlated_features}")

```

```python

import pandas as pd

# حساب إجمالي الأصول
df['total_assets'] = df['gained_asset'] + df['durable_asset'] + df['save_asset']

# حساب إجمالي النفقات
df['total_expenses'] = df[['farm_expenses', 'living_expenses', 'other_expenses']].sum(axis=1)

# عرض البيانات بشكل مختصر مع الاحتفاظ بالفهارس الأصلية
first_five = df[['total_assets', 'total_expenses']].head()
last_five = df[['total_assets', 'total_expenses']].tail()

# دمج البيانات مع الصف الخاص بـ "..." مع الحفاظ على الفهارس الأصلية للصفوف الأخيرة
summary_df = pd.concat([first_five, 
                        pd.DataFrame([['...', '...']], columns=['total_assets', 'total_expenses'], index=['...']), 
                        last_five])

# عرض الجدول المختصر مع الفهارس الأصلية
print(summary_df)
```

```python
import matplotlib.pyplot as plt
# Define outliers based on IQR
Q1 = df['Age'].quantile(0.25)
Q3 = df['Age'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter outliers
outliers = df[(df['Age'] < lower_bound) | (df['Age'] > upper_bound)]

# Create scatter plot with highlighted outliers
plt.figure(figsize=(10, 6))
plt.scatter(df['Age'], df['depressed'], color='blue', alpha=0.5, label='Normal Data')
plt.scatter(outliers['Age'], outliers['depressed'], color='red', label='Outliers')
plt.title('Scatter Plot of Age vs. Depression Score with Outliers Highlighted')
plt.xlabel('Age')
plt.ylabel('depressed')
plt.legend()
plt.grid(True)
plt.show()
```

```python
print("The age after discretization")

# تقسيم الأعمار إلى فئات
age_bins = [0, 25, 60, df['Age'].max()]  # الفواصل الزمنية لتقسيم الأعمار
age_labels = ['Youth', 'Adult', 'Senior']  # تسميات الفئات العمرية

# إنشاء عمود جديد 'Age_Group' بناءً على التقسيم
df['Age_Group'] = pd.cut(df['Age'], bins=age_bins, labels=age_labels, include_lowest=True)

# عرض أول 5 صفوف وآخر 5 صفوف مع الحفاظ على الفهارس الأصلية
first_five = df[['Age', 'Age_Group']].head()
last_five = df[['Age', 'Age_Group']].tail()

# دمج البيانات مع صف "..." في المنتصف للحفاظ على الترتيب الأصلي
summary_df = pd.concat([first_five, 
                        pd.DataFrame([['...', '...']], columns=['Age', 'Age_Group'], index=['...']),
                        last_five])

# عرض الجدول المختصر مع الفهارس الأصلية
print(summary_df)
```
